```python
import numpy as np
import pandas as pd
```

```python
inc_08 = pd.read_stata('ci08a_1.0p_EN.dta')
inc_10 = pd.read_stata('ci10c_EN_1.0p.dta')
inc_12 = pd.read_stata('ci12e_1.0p_EN.dta')
inc_14 = pd.read_stata('ci14g_1.0p_EN.dta')
inc_16 = pd.read_stata('ci16i_EN_3.0p.dta')
inc_18 = pd.read_stata('ci18k_EN_2.0p.dta')
inc_20 = pd.read_stata('ci20m_EN_1.0p.dta')
inc_22 = pd.read_stata('ci22o_EN_1.0p.dta')
```

```python
inc_all = [inc_08, inc_10, inc_12, inc_14, inc_16, inc_18, inc_20, inc_22]

inc_merged = inc_all[0]

for inc in inc_all[1:]:
    inc_merged = pd.merge(inc_merged, inc, on='nomem_encr', how='inner')
```
